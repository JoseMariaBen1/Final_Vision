```python
import cv2
print("OpenCV should be 4.8.0.76 Current version:", cv2.__version__)
from typing import List
import numpy as np
import imageio
import cv2
import copy
import glob
```

```python
def load_images(filenames: List) -> List:
    return [imageio.imread(filename) for filename in filenames]
```

```python
# TODO Build a list containing the paths of all images from the left camera
imgs_path = [f"Calibration_images/{i:01d}.jpg" for i in range(1,12)]
imgs = load_images(imgs_path)
```

```python
corners_img = []
for img in imgs:
    corners = cv2.findChessboardCorners(img, patternSize = (9,6))
    corners_img.append(corners)

```

```python

corners_copies = []

for corners in corners_img:
    corners_copy = copy.deepcopy(corners)
    corners_copies.append(corners_copy)


criteria = (cv2.TERM_CRITERIA_EPS + cv2.TERM_CRITERIA_MAX_ITER, 30, 0.01)

# TODO To refine corner detections with cv2.cornerSubPix() you need to input grayscale images. Build a list containing grayscale images.
imgs_gray = []
for img in imgs:
    img_gray = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
    imgs_gray.append(img_gray)



corners_refined = [cv2.cornerSubPix(i, cor[1], (8, 6), (-1, -1), criteria) if cor[0] else [] for i, cor in zip(imgs_gray, corners_copies)]
```
