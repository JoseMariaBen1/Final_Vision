### **Tarea A.4:** Cambia el tamaÃ±o de una imagen ``cv2.resize()`` a 200x200 pÃ­xeles.

```python
# TODO: Resize the image to 200Ã—200 pixels using cv2.resize().
resized = cv2.resize(img,[200,200])
# TODO: Check the image shape
print(f"Resized shape:{resized.shape}", ...)
# TODO: Display the image using cv2.imshow().
cv2.imshow('Image_resized',resized)
cv2.waitKey()
cv2.destroyAllWindows()
```

### **Tarea A.5:** Recorta el centro de la imagen a un tamaÃ±o 120x160.

```python
# TODO: # TODO: Crop the center of the image 'img' with an approximate size of 120x160 pixels.
h, w = img.shape[:2]
cy, cx = h//2, w//2
roi = img[cy-60:cy+60,cx-80:cx+80]
# TODO: Check the image shape
print(f"ROI shape: {roi.shape}", ...)
# TODO: Display the image using cv2.imshow().
cv2.imshow('Image_ROI',roi)
cv2.waitKey()
cv2.destroyAllWindows()
```

```python
# TODO: Rotate around the center using cv2.getRotationMatrix2D and cv2.warpAffine.
M = cv2.getRotationMatrix2D((h/2,w/2),angle=45,scale=1)
rotated = cv2.warpAffine(img,M, img.shape[:2])
# TODO: Display the image using cv2.imshow().
cv2.imshow('Image_Rotated',rotated)
cv2.waitKey()
cv2.destroyAllWindows()
```

### **Tarea B.1:** Defina y ejecute el mÃ©todo para cargar imÃ¡genes ``load_images()``

```python
def load_images(filenames: List) -> List:
    return [cv2.imread(filename) for filename in filenames]
```

```python
# TODO Build a list containing the paths of all images in the data folder

imgs_path = [f'{data_dir}/00{i}.jpg' for i in range(6)]

imgs = load_images(imgs_path)
```

### **Tarea B.2**. Defina los mÃ©todos ``show_image()`` y ``write_image()`` para visualizar y guardar imÃ¡genes

```python
# TODO Complete the method, use every argument
def show_image(img: np.array, img_name: str = "Image"):
    cv2.imshow(img_name,img)
    cv2.waitKey()
    cv2.destroyAllWindows()

# TODO Complete the method, use every argument
def write_image(output_folder: str, img_name: str, img: np.array):
    img_path = os.path.join(output_folder,img_name)
    cv2.imwrite(img_path,img)
```

### **Tarea B.3:** Cambie el espacio de color de las imÃ¡genes a uno donde la crominancia y la intensidad estÃ©n separados (HSV)

```python
# TODO Get a list with Ã¬mgs in HSV color space
imgs_hsv = [cv2.cvtColor(img, cv2.COLOR_BGR2HSV) for img in imgs]
```

### **Tarea B.4:** Segmente los colores anaranjados

```python
# TODO Define orange color range
light_orange = (1, 190, 200)
dark_orange = (255, 255, 255)

# TODO Compute a list that contains a mask (which should segment orange colors) for every image.

orange_masks = [cv2.inRange(img, light_orange,dark_orange)for img in imgs_hsv]

# TODO Compute a list that contains the result of multiplying the original image with its orange colors mask.
orange_segmented = [cv2.bitwise_and(imgs_hsv[i],imgs_hsv[i],mask=orange_masks[i]) for i in range(len(imgs_hsv))]

# TODO Show an original image
show_image(cv2.cvtColor(imgs_hsv[0], cv2.COLOR_HSV2BGR))

# TODO Show a mask
show_image(orange_masks[0])

# TODO Show a segmented image
show_image(cv2.cvtColor(orange_segmented[0], cv2.COLOR_HSV2BGR))
```

### **Tarea B.5:** Segmente los colores blancos.

Para detectar el rango de blancos complete la siguiente celda y ejecutela para investigar el rango de valores necesarios.

```python
# TODO Discover white color ranges
#get_hsv_color_ranges(img)

# TODO Define white color range
light_white = (0, 0, 200)
dark_white = (145, 60, 255)

# TODO Compute a list that contains a mask (which should segment white colors) for every image.
white_masks = [cv2.inRange(img, light_white,dark_white)for img in imgs_hsv]

# TODO Compute a list that contains the result of multiplying the original image with its white colors mask.
white_segmented = [cv2.bitwise_and(imgs_hsv[i],imgs_hsv[i],mask=white_masks[i]) for i in range(len(imgs_hsv))]


# TODO Show an original image
show_image(cv2.cvtColor(imgs_hsv[0], cv2.COLOR_HSV2BGR))

# TODO Show a mask
show_image(white_masks[0])

# TODO Show a segmented image
show_image(cv2.cvtColor(white_segmented[0], cv2.COLOR_HSV2BGR))
```

### **Tarea B.6:** Junte las mÃ¡scaras para cada imagen (naranja + blanco) y segmente cada una de ellas.

```python
# TODO Join orange_masks and white_masks
fish_masks = [cv2.bitwise_or(white_masks[i], orange_masks[i]) for i in range(len(imgs))]

    
# TODO Compute a list that contains the result of multiplying the original image with its complete mask.
fish = [cv2.bitwise_and(imgs_hsv[i],imgs_hsv[i],mask=fish_masks[i]) for i in range(len(imgs_hsv))]

# TODO Show an original image
show_image(cv2.cvtColor(imgs_hsv[0], cv2.COLOR_HSV2BGR))


# TODO Show a mask
show_image(fish_masks[0])

# TODO Show a segmented image
show_image(cv2.cvtColor(fish[0], cv2.COLOR_HSV2BGR))
```

### **Tarea B.7:** Guarde las imÃ¡genes

```python
# TODO Define your output folder and save every fish image
output_folder = "../Parte1B"
os.makedirs(output_folder, exist_ok=True)

for i in range(len(imgs_hsv)):
    write_image(output_folder=output_folder, img_name=f"Image_00{i}_hsv.jpg", img=imgs_hsv[i])

for i in range(len(orange_masks)):
    write_image(output_folder=output_folder, img_name=f"Image_00{i}_orange_mask.jpg", img=orange_masks[i])

for i in range(len(orange_segmented)):
    write_image(output_folder=output_folder, img_name=f"Image_00{i}_orange_segmented.jpg", img=cv2.cvtColor(orange_segmented[i], cv2.COLOR_HSV2BGR))

for i in range(len( white_masks)):
    write_image(output_folder=output_folder, img_name=f"Image_00{i}_white_mask.jpg", img=white_masks[i])

for i in range(len(white_segmented)):
    write_image(output_folder=output_folder, img_name=f"Image_00{i}_white_segmented.jpg", img=(cv2.cvtColor(white_segmented[i], cv2.COLOR_HSV2BGR)))

for i in range(len(fish_masks)):
    write_image(output_folder=output_folder, img_name=f"Image_00{i}_fish_mask.jpg", img=fish_masks[i])

for i in range(len(fish)):
    write_image(output_folder=output_folder, img_name=f"Image_00{i}_fish.jpg", img=(cv2.cvtColor(fish[i], cv2.COLOR_HSV2BGR)))

```
